```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sfs
#plt.rcParams['figure.figsize'] = 8, 4.5  # inch

# Settings
X = [-2, 2]
Y = [-2, 2]
Z = 0
pw_angle = 45  # traveling direction of plane wave
f = 800  # frequency

# Calculations
grid = sfs.util.xyz_grid(X, Y, Z, spacing=0.02)
n0 = sfs.util.direction_vector(np.radians(pw_angle))
x0 = [0, 0, 0]
omega = 2 * np.pi * f
p = sfs.mono.source.plane(omega, x0, n0, grid)
sfs.plot.soundfield(p, grid);
```
